## 2.1 Características generales

Un **criterio de plastificación** generaliza el concepto de límite elástico al caso
multiaxial. Siendo $\sigma$ el estado tensional:

$$f(\sigma)<0\;\Rightarrow\;\text{régimen elástico},\qquad
  f(\sigma)=0\;\Rightarrow\;\text{régimen plástico}$$

**Hipótesis: isotropía.** El criterio solo puede depender de los **invariantes** de
$\sigma$ (equivalentemente, de las tensiones principales $\sigma_1,\sigma_2,\sigma_3$):

$$J_1=\sigma_1+\sigma_2+\sigma_3=\operatorname{tr}\sigma$$

Descomposición en parte **esférica** y **desviadora**,
$\;\sigma=\sigma'+\tfrac13(\operatorname{tr}\sigma)\,\mathbf 1$, con los invariantes del
desviador:

$$J_2'=\tfrac12\,\sigma'\!:\!\sigma',\qquad J_3'=\det\sigma'$$

De forma general, un criterio isótropo se expresa como $\;f(J_2',J_3')=0$.

**La plastificación no depende de la presión hidrostática:** una compresión igual en todas
las direcciones solo cambia el *volumen*, no la *forma*. Por eso los criterios se escriben
con los invariantes del **desviador** ($J_2'$, $J_3'$) y **no** con $J_1$.

> **Consecuencia geométrica** (clave para las figuras de abajo): en el espacio de tensiones
> principales las superficies de fluencia son **prismas/cilindros infinitos** cuyo eje es la
> **recta hidrostática** $\sigma_1=\sigma_2=\sigma_3$. Su sección perpendicular a ese eje
> (el **plano desviador** o **plano $\pi$**) es lo que distingue a un criterio de otro.

## 2.2 Criterios de Tresca y Von Mises

### Tresca (tensión tangencial máxima)

Plastifica cuando la **tensión tangencial máxima** alcanza un valor crítico $k$. Ordenando
$\sigma_1\ge\sigma_2\ge\sigma_3$:

$$\tau_{\max}=\frac{\sigma_1-\sigma_3}{2}=k
\qquad\Longrightarrow\qquad\boxed{\;\sigma_1-\sigma_3=Y\;}$$

**Calibración con el ensayo de tracción simple**, donde $(\sigma_1,\sigma_2,\sigma_3)=(Y,0,0)$:

$$\frac{Y-0}{2}=k\qquad\Longrightarrow\qquad
\boxed{\;k_{\text{Tresca}}=\frac{Y}{2}=0{,}500\,Y\;}$$

En función de los invariantes del desviador:

$$4\left(J_2'-k^2\right)\left(J_2'-4k^2\right)^2=27\,J_3'^{\,2}$$

### Von Mises (1913)

$$\boxed{\;J_2'=k^2\;}$$

con

$$J_2'=\tfrac12\,\sigma'\!:\!\sigma'
=\tfrac12\left(\sigma_1'^{\,2}+\sigma_2'^{\,2}+\sigma_3'^{\,2}\right)
=\tfrac16\left[(\sigma_1-\sigma_2)^2+(\sigma_2-\sigma_3)^2+(\sigma_3-\sigma_1)^2\right]$$

o, en ejes generales,

$$J_2'=\tfrac16\left[(\sigma_x-\sigma_y)^2+(\sigma_y-\sigma_z)^2+(\sigma_z-\sigma_x)^2
+6\left(\tau_{xy}^2+\tau_{xz}^2+\tau_{yz}^2\right)\right]$$

Se define la **tensión equivalente de Von Mises**:

$$\bar\sigma=\sqrt{3J_2'}
=\sqrt{\tfrac12\left[(\sigma_1-\sigma_2)^2+(\sigma_2-\sigma_3)^2+(\sigma_3-\sigma_1)^2\right]}
\qquad\Longrightarrow\qquad \bar\sigma=Y\ \text{en fluencia}$$

**Calibración con el ensayo de tracción simple** $(\sigma_1,\sigma_2,\sigma_3)=(Y,0,0)$:

$$J_2'=\tfrac16\left(Y^2+0+Y^2\right)=\frac{Y^2}{3}=k^2
\qquad\Longrightarrow\qquad\boxed{\;k_{\text{Mises}}=\frac{Y}{\sqrt3}\approx0{,}577\,Y\;}$$

### Comparación

Ambos criterios se calibran para **coincidir en tracción simple**, pero difieren en el resto
de estados. La discrepancia **máxima** aparece en **cortante puro** $(k,0,-k)$:

| criterio | $k$ en cortante puro | |
|---|---|---|
| **Tresca** | $Y/2=0{,}500\,Y$ | más **conservador** |
| **Von Mises** | $Y/\sqrt3\approx0{,}577\,Y$ | |

Von Mises admite $\;2/\sqrt3\approx1{,}155\;$ veces más tensión tangencial que Tresca, es
decir un **15,5 % más**. Geométricamente **Tresca queda inscrito en Von Mises** y ambos
**coinciden en 6 puntos** (los estados uniaxiales).

> **Nota sobre $Y$.** Aquí $Y$ es el límite elástico del material (medido en el ensayo de
> tracción, §1). Con **endurecimiento isótropo** (§3.1) $Y$ crece con la deformación
> plástica acumulada y ambas superficies se **expanden homotéticamente**, sin cambiar de
> forma: mover el deslizador $Y$ reproduce justamente ese efecto.

```python
# ---------- tensiones equivalentes ----------
def von_mises_3d(s1, s2, s3):
    """Tensión equivalente de Von Mises a partir de las tensiones principales."""
    return np.sqrt(0.5*((s1 - s2)**2 + (s2 - s3)**2 + (s3 - s1)**2))


def tresca_3d(s1, s2, s3):
    """Tensión equivalente de Tresca: sigma_max - sigma_min."""
    return max(s1, s2, s3) - min(s1, s2, s3)


def von_mises_plane_stress(sxx, syy, txy):
    """Von Mises en tensión plana (sigma_3 = 0)."""
    return np.sqrt(sxx**2 - sxx*syy + syy**2 + 3.0*txy**2)


def principal_stresses(sxx, syy, txy):
    """Tensiones principales en el plano (círculo de Mohr)."""
    avg = 0.5*(sxx + syy)
    R = np.sqrt((0.5*(sxx - syy))**2 + txy**2)
    return avg + R, avg - R


# ---------- lugares geométricos en TENSIÓN PLANA (sigma_3 = 0) ----------
def yield_ellipse(Y, n=400):
    """Von Mises: sigma_1^2 - sigma_1 sigma_2 + sigma_2^2 = Y^2  (elipse)."""
    t = np.linspace(0.0, 2.0*np.pi, n)
    u = np.sqrt(2.0)*Y*np.cos(t)        # eje mayor, a lo largo de sigma_1 = sigma_2
    v = np.sqrt(2.0/3.0)*Y*np.sin(t)    # eje menor, a lo largo de sigma_1 = -sigma_2
    return (u + v)/np.sqrt(2.0), (u - v)/np.sqrt(2.0)


def tresca_hexagon(Y):
    """Tresca en tensión plana: max(|s1-s2|,|s1|,|s2|) = Y  (hexágono)."""
    v = np.array([[Y, 0], [Y, Y], [0, Y], [-Y, 0], [-Y, -Y], [0, -Y], [Y, 0]], float)
    return v[:, 0], v[:, 1]


# ---------- plano desviador (plano pi) ----------
_n  = np.array([1.0, 1.0, 1.0])/np.sqrt(3.0)     # eje hidrostático
_e1 = np.array([1.0, -1.0, 0.0])/np.sqrt(2.0)    # base ortonormal del plano pi
_e2 = np.array([1.0, 1.0, -2.0])/np.sqrt(6.0)


def pi_project(p):
    """Proyecta un punto (s1,s2,s3) sobre el plano desviador."""
    p = np.asarray(p, float)
    dev = p - p.sum()/3.0
    return float(dev @ _e1), float(dev @ _e2)


def pi_circle(Y, n=400):
    """Von Mises en el plano pi: circunferencia de radio sqrt(2/3) Y."""
    t = np.linspace(0.0, 2.0*np.pi, n)
    R = np.sqrt(2.0/3.0)*Y
    return R*np.cos(t), R*np.sin(t)


def pi_hexagon(Y):
    """Tresca en el plano pi: hexágono regular inscrito (vértices = estados uniaxiales)."""
    states = [(Y, 0, 0), (0, Y, 0), (0, 0, Y), (-Y, 0, 0), (0, -Y, 0), (0, 0, -Y)]
    pts = np.array([pi_project(s) for s in states])
    pts = pts[np.argsort(np.arctan2(pts[:, 1], pts[:, 0]))]
    pts = np.vstack([pts, pts[0]])
    return pts[:, 0], pts[:, 1]
```

## Sección en **tensión plana** ($\sigma_3=0$): hexágono vs. elipse

Haciendo $\sigma_3=0$ en ambos criterios se obtienen los lugares geométricos del plano
$(\sigma_1,\sigma_2)$:

**Von Mises** $\;\rightarrow\;$ **elipse**

$$\sigma_1^2-\sigma_1\sigma_2+\sigma_2^2=Y^2$$

de semiejes $\sqrt2\,Y$ (según $\sigma_1=\sigma_2$) y $\sqrt{2/3}\,Y$ (según
$\sigma_1=-\sigma_2$).

**Tresca** $\;\rightarrow\;$ **hexágono**

$$\max\left(|\sigma_1-\sigma_2|,\;|\sigma_1|,\;|\sigma_2|\right)=Y$$

de vértices $(Y,0)$, $(Y,Y)$, $(0,Y)$, $(-Y,0)$, $(-Y,-Y)$, $(0,-Y)$ — que son
precisamente los **6 puntos donde ambos criterios coinciden**.

Mueve el punto de tensión con $\sigma_{xx},\sigma_{yy},\tau_{xy}$ (se calculan las
principales con el círculo de Mohr). El punto se colorea:

* 🟢 **verde** — elástico según **los dos** criterios,
* 🟠 **naranja** — en la **banda entre hexágono y elipse**: Tresca dice que plastifica,
  Von Mises dice que no,
* 🔴 **rojo** — plastifica según **los dos**.

```python
# --- figura PERSISTENTE ---
plt.ioff()
_limA = {}
figA, axA = plt.subplots(figsize=(7.2, 7.2))
plt.ion()


def plot_plane_stress(Y=250.0, sxx=0.0, syy=0.0, txy=135.0):
    ax = axA
    ax.clear()

    s1e, s2e = yield_ellipse(Y)          # Von Mises
    hx, hy = tresca_hexagon(Y)           # Tresca
    s1, s2 = principal_stresses(sxx, syy, txy)

    svm = von_mises_plane_stress(sxx, syy, txy)      # = von_mises_3d(s1, s2, 0)
    st = tresca_3d(s1, s2, 0.0)

    in_vm, in_tr = svm <= Y, st <= Y
    if in_vm and in_tr:
        col, estado = 'seagreen', 'ELÁSTICO según ambos criterios'
    elif in_vm and not in_tr:
        col, estado = 'darkorange', 'PLASTIFICA según Tresca  |  ELÁSTICO según Von Mises'
    else:
        col, estado = 'crimson', 'PLASTIFICA según ambos criterios'

    # superficies
    ax.plot(s1e, s2e, color=C_VM, lw=2.2, label=r'Von Mises  $\bar\sigma=Y$')
    ax.fill(s1e, s2e, color=C_VM, alpha=0.06)
    ax.plot(hx, hy, color=C_TR, lw=2.2, label=r'Tresca  $\sigma_1-\sigma_3=Y$')
    ax.fill(hx, hy, color=C_TR, alpha=0.06)

    # ejes y recta hidrostática
    lim = max(np.abs(s1e).max(), np.abs(s2e).max())*1.18
    ax.axhline(0, color='0.6', lw=0.8); ax.axvline(0, color='0.6', lw=0.8)
    ax.plot([-lim, lim], [-lim, lim], ls='--', color='0.75', lw=0.8,
            label=r'$\sigma_1=\sigma_2$')

    # 6 puntos de coincidencia
    ax.plot(hx[:-1], hy[:-1], 'o', color='0.25', ms=4, zorder=6,
            label='coinciden (6 puntos)')

    # punto de tensión
    ax.plot([0, s1], [0, s2], ls=':', color=col, lw=1.2)
    ax.plot(s1, s2, 'o', color=col, ms=12, zorder=7,
            markeredgecolor='k', markeredgewidth=0.6)
    ax.annotate(r'$(%.0f,\;%.0f)$' % (s1, s2), (s1, s2), textcoords='offset points',
                xytext=(10, -4), color=col, fontsize=10)

    fix_limits(_limA, ax, (-lim, lim), (-lim, lim))
    ax.set_aspect('equal')
    ax.set_xlabel(r'$\sigma_1$  (MPa)')
    ax.set_ylabel(r'$\sigma_2$  (MPa)')
    ax.grid(alpha=0.3)
    ax.legend(loc='upper left', fontsize=9)
    _t = r'$\bar\sigma_{VM}=%.0f$   |   $\sigma_{Tresca}=%.0f$   |   $Y=%.0f$ MPa'
    ax.set_title(_t % (svm, st, Y) + '\n' + estado, color=col, fontsize=11)
    figA.canvas.draw_idle()


panel(plot_plane_stress, dict(
    Y=FloatSlider(value=250, min=50, max=600, step=10,
                  description='límite elást. Y (MPa)', style={'description_width':'initial'}),
    sxx=FloatSlider(value=0, min=-600, max=600, step=10,
                    description='σ_xx (MPa)', style={'description_width':'initial'}),
    syy=FloatSlider(value=0, min=-600, max=600, step=10,
                    description='σ_yy (MPa)', style={'description_width':'initial'}),
    txy=FloatSlider(value=135, min=-400, max=400, step=5,
                    description='τ_xy (MPa)', style={'description_width':'initial'}),
), figA)
```

## El **plano desviador** (plano $\pi$): círculo vs. hexágono

Mirando el espacio de tensiones principales **a lo largo de la recta hidrostática**
($\sigma_1=\sigma_2=\sigma_3$) se ve la sección que define cada criterio. Es la vista que
mejor los compara, porque elimina la parte esférica:

* **Von Mises** $\rightarrow$ **circunferencia** de radio
  $\;\rho=\sqrt{\tfrac23}\,Y\approx0{,}8165\,Y$,
* **Tresca** $\rightarrow$ **hexágono regular inscrito**, cuyos **vértices** son los estados
  uniaxiales (allí coinciden) y cuyos **lados planos** están a la distancia (apotema)
  $\;\rho\cos 30^\circ=Y/\sqrt2\approx0{,}7071\,Y$.

La separación es **nula en los vértices** (tracción/compresión simple) y **máxima en el
centro de los lados** (cortante puro), donde

$$\frac{\rho_{\text{Mises}}}{\rho_{\text{Tresca}}}=\frac{1}{\cos 30^\circ}=\frac{2}{\sqrt3}
\approx1{,}155\qquad(\textbf{15,5 \%})$$

```python
# --- figura PERSISTENTE ---
plt.ioff()
_limB = {}
figB, axB = plt.subplots(figsize=(7.0, 7.0))
plt.ion()


def plot_pi_plane(Y=250.0):
    ax = axB
    ax.clear()

    cx, cy = pi_circle(Y)
    hx, hy = pi_hexagon(Y)
    R = np.sqrt(2.0/3.0)*Y          # radio Von Mises
    a = R*np.cos(np.pi/6.0)         # apotema del hexágono = Y/sqrt(2)

    ax.plot(cx, cy, color=C_VM, lw=2.2, label=r'Von Mises  ($\rho=\sqrt{2/3}\,Y$)')
    ax.fill(cx, cy, color=C_VM, alpha=0.05)
    ax.plot(hx, hy, color=C_TR, lw=2.2, label=r'Tresca (hexágono inscrito)')
    ax.fill(hx, hy, color=C_TR, alpha=0.07)

    # vértices = estados uniaxiales -> coinciden
    ax.plot(hx[:-1], hy[:-1], 'o', color='0.25', ms=5, zorder=6,
            label='uniaxial: coinciden')

    # proyección de los ejes principales
    for vec, lab in [((1, 0, 0), r'$\sigma_1$'), ((0, 1, 0), r'$\sigma_2$'),
                     ((0, 0, 1), r'$\sigma_3$')]:
        x, y = pi_project(np.array(vec)*R*1.55)
        ax.plot([0, x], [0, y], color='0.7', lw=0.8, ls='-')
        ax.annotate(lab, (x, y), textcoords='offset points', xytext=(4, 3),
                    fontsize=11, color='0.45')

    # cortante puro: centro de un lado -> separación máxima (15,5 %)
    xm, ym = 0.5*(hx[0] + hx[1]), 0.5*(hy[0] + hy[1])
    u = np.hypot(xm, ym)
    ax.plot([xm/u*a, xm/u*R], [ym/u*a, ym/u*R], color='k', lw=1.6, zorder=7)
    ax.plot(xm/u*a, ym/u*a, 's', color=C_TR, ms=6, zorder=7)
    ax.plot(xm/u*R, ym/u*R, 's', color=C_VM, ms=6, zorder=7)
    ax.annotate('cortante puro\n15,5 %', (xm/u*(a + R)/2, ym/u*(a + R)/2),
                textcoords='offset points', xytext=(12, -6), fontsize=10)

    lim = R*1.75
    fix_limits(_limB, ax, (-lim, lim), (-lim, lim))
    ax.set_aspect('equal')
    ax.set_xlabel('plano desviador  (MPa)')
    ax.set_ylabel('plano desviador  (MPa)')
    ax.grid(alpha=0.3)
    ax.legend(loc='upper left', fontsize=9)
    _t = r'Plano $\pi$:  $\rho_{Mises}=%.1f$   vs   apotema$_{Tresca}=%.1f$ MPa'
    ax.set_title(_t % (R, a) + '\n' + r'(vista a lo largo de $\sigma_1=\sigma_2=\sigma_3$)',
                 fontsize=11)
    figB.canvas.draw_idle()


panel(plot_pi_plane, dict(
    Y=FloatSlider(value=250, min=50, max=600, step=10,
                  description='límite elást. Y (MPa)', style={'description_width':'initial'}),
), figB)
```

## Vista 3D: **cilindro** (Von Mises) vs. **prisma hexagonal** (Tresca)

Como ningún criterio depende de la presión hidrostática, ambas superficies son
**infinitas** en la dirección $\sigma_1=\sigma_2=\sigma_3$:

* **Von Mises** $\rightarrow$ **cilindro circular** de radio $\sqrt{2/3}\,Y$,
* **Tresca** $\rightarrow$ **prisma hexagonal** inscrito en él.

Todo lo que queda **dentro** es régimen elástico. Las secciones que ya has visto son
cortes de estos volúmenes: el **plano $\pi$** es el corte perpendicular al eje, y la
**tensión plana** es el corte con $\sigma_3=0$.

* **Gira con el ratón**, rueda para hacer zoom.
* Los deslizadores cambian $Y$ y el punto de tensión $(\sigma_1,\sigma_2,\sigma_3)$, que se
  colorea con el mismo código que antes (verde / naranja / rojo).
* Girar la vista **no se reinicia** al mover un deslizador.

```python
def vm_cylinder(Y, half_len, n_theta=64):
    """Malla del cilindro de Von Mises (radio sqrt(2/3) Y) en torno al eje hidrostático."""
    rho = np.sqrt(2.0/3.0)*Y
    th = np.linspace(0.0, 2.0*np.pi, n_theta)
    h = np.array([-half_len, half_len])
    TH, HH = np.meshgrid(th, h)
    X = HH*_n[0] + rho*(np.cos(TH)*_e1[0] + np.sin(TH)*_e2[0])
    Y3 = HH*_n[1] + rho*(np.cos(TH)*_e1[1] + np.sin(TH)*_e2[1])
    Z = HH*_n[2] + rho*(np.cos(TH)*_e1[2] + np.sin(TH)*_e2[2])
    return X, Y3, Z


def tresca_prism(Y, half_len):
    """Malla del prisma hexagonal de Tresca (mismo eje, inscrito en el cilindro)."""
    hx, hy = pi_hexagon(Y)                     # 7 puntos (cerrado)
    h = np.array([-half_len, half_len])
    X = np.zeros((2, len(hx))); Y3 = np.zeros_like(X); Z = np.zeros_like(X)
    for j, hh in enumerate(h):
        for i in range(len(hx)):
            P = hh*_n + hx[i]*_e1 + hy[i]*_e2
            X[j, i], Y3[j, i], Z[j, i] = P
    return X, Y3, Z


def _axis_line(vec, L):
    v = np.asarray(vec, float)
    return dict(x=[-L*v[0], L*v[0]], y=[-L*v[1], L*v[1]], z=[-L*v[2], L*v[2]])


# ---- deslizadores ----
_w = dict(style={'description_width': 'initial'}, continuous_update=False)
sl_Y  = FloatSlider(value=250, min=50, max=600, step=10, description='límite elást. Y (MPa)', **_w)
sl_s1 = FloatSlider(value=140, min=-600, max=600, step=10, description='σ_1 (MPa)', **_w)
sl_s2 = FloatSlider(value=0, min=-600, max=600, step=10, description='σ_2 (MPa)', **_w)
sl_s3 = FloatSlider(value=-140, min=-600, max=600, step=10, description='σ_3 (MPa)', **_w)

_L0 = 2.6*sl_Y.value                      # semilongitud fija de los volúmenes
X0, Y0, Z0 = vm_cylinder(sl_Y.value, _L0)
Xp, Yp, Zp = tresca_prism(sl_Y.value, _L0)

fig3d = go.FigureWidget(layout=go.Layout(
    width=780, height=640, margin=dict(l=0, r=0, t=44, b=0), font=PLOTLY_FONT,
    scene=dict(xaxis_title='σ1 (MPa)', yaxis_title='σ2 (MPa)', zaxis_title='σ3 (MPa)',
               aspectmode='cube', camera=dict(eye=dict(x=1.5, y=1.5, z=1.2))),
    title='Von Mises (cilindro) vs Tresca (prisma hexagonal)'))

# 0: cilindro Von Mises
fig3d.add_surface(x=X0, y=Y0, z=Z0, opacity=0.28, showscale=False,
                  colorscale=[[0, C_VM], [1, C_VM]], name='Von Mises', hoverinfo='skip')
# 1: prisma Tresca
fig3d.add_surface(x=Xp, y=Yp, z=Zp, opacity=0.45, showscale=False,
                  colorscale=[[0, C_TR], [1, C_TR]], name='Tresca', hoverinfo='skip')
# 2: eje hidrostático
_ah = _axis_line(_n, _L0)
fig3d.add_scatter3d(x=_ah['x'], y=_ah['y'], z=_ah['z'], mode='lines',
                    line=dict(color='black', width=4, dash='dash'),
                    name='eje hidrostático', hoverinfo='skip')
# 3,4,5: ejes principales
for _v, _nm, _c in [((1, 0, 0), 'σ1', 'firebrick'), ((0, 1, 0), 'σ2', 'green'),
                    ((0, 0, 1), 'σ3', 'royalblue')]:
    _a = _axis_line(_v, _L0)
    fig3d.add_scatter3d(x=_a['x'], y=_a['y'], z=_a['z'], mode='lines',
                        line=dict(color=_c, width=2), name=_nm, hoverinfo='skip')
# 6: punto de tensión
fig3d.add_scatter3d(x=[sl_s1.value], y=[sl_s2.value], z=[sl_s3.value], mode='markers',
                    marker=dict(size=6, color='seagreen'), name='punto de tensión')


def _update3d(change=None):
    Y = sl_Y.value
    s1, s2, s3 = sl_s1.value, sl_s2.value, sl_s3.value
    svm = von_mises_3d(s1, s2, s3)
    st = tresca_3d(s1, s2, s3)
    in_vm, in_tr = svm <= Y, st <= Y
    if in_vm and in_tr:
        col, estado = 'seagreen', 'ELÁSTICO según ambos'
    elif in_vm and not in_tr:
        col, estado = 'darkorange', 'PLASTIFICA solo según Tresca'
    else:
        col, estado = 'crimson', 'PLASTIFICA según ambos'
    Xc, Yc, Zc = vm_cylinder(Y, _L0)
    Xt, Yt, Zt = tresca_prism(Y, _L0)
    with fig3d.batch_update():
        fig3d.data[0].x, fig3d.data[0].y, fig3d.data[0].z = Xc, Yc, Zc
        fig3d.data[1].x, fig3d.data[1].y, fig3d.data[1].z = Xt, Yt, Zt
        fig3d.data[6].x, fig3d.data[6].y, fig3d.data[6].z = [s1], [s2], [s3]
        fig3d.data[6].marker.color = col
        fig3d.layout.title = (f'Y={Y:.0f}   |   σ_VM={svm:.0f}   |   σ_Tresca={st:.0f} MPa'
                              f'   →   {estado}')


for _s in (sl_Y, sl_s1, sl_s2, sl_s3):
    _s.observe(_update3d, names='value')
_update3d()

display(widgets.HBox([
    widgets.VBox([widgets.HTML('<b>Material</b>'), sl_Y,
                  widgets.HTML('<b>Punto de tensión</b>'), sl_s1, sl_s2, sl_s3]),
    fig3d]))
```

## Resumen

| | **Tresca** | **Von Mises** |
|---|---|---|
| criterio | $\sigma_1-\sigma_3=Y$ | $J_2'=k^2$, $\;\bar\sigma=\sqrt{3J_2'}=Y$ |
| base física | tensión tangencial máxima | energía de distorsión ($J_2'$) |
| $k$ (cortante puro) | $Y/2=0{,}500\,Y$ | $Y/\sqrt3\approx0{,}577\,Y$ |
| tensión plana | hexágono | elipse |
| plano $\pi$ | hexágono regular inscrito | circunferencia $\rho=\sqrt{2/3}\,Y$ |
| espacio 3D | prisma hexagonal | cilindro circular |
| carácter | **más conservador** | más ajustado a los ensayos en metales dúctiles |

* Ninguno depende de la **presión hidrostática** $\Rightarrow$ superficies **infinitas** con
  eje en la recta hidrostática.
* Ambos se **calibran con el ensayo de tracción simple** (§1) y **coinciden** en los 6
  estados uniaxiales.
* Diferencia **máxima del 15,5 %** ($2/\sqrt3$) en **cortante puro**.
* Para materiales con distinta resistencia a tracción y compresión (suelos, hormigón,
  rocas) se emplea **Mohr–Coulomb** (§2.3), que **sí** depende de la presión.
